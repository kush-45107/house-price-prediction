# Model 

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load Excel file
df = pd.read_excel("Dataset.xlsx")

# Drop the 'Dataset' column if it's not needed
df.drop(columns=['Dataset'], inplace=True)

# Columns to encode
categorical_cols = ['Location', 'Parking', 'Property_type', 'Furnishing_status']

# Apply OneHotEncoder
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_array = encoder.fit_transform(df[categorical_cols])

# Get encoded column names
encoded_cols = encoder.get_feature_names_out(categorical_cols)

# Create a DataFrame with encoded features
encoded_df = pd.DataFrame(encoded_array, columns=encoded_cols, index=df.index)

# Combine with rest of the data
df_encoded = df.drop(columns=categorical_cols).join(encoded_df)

# Save the encoded dataset
df_encoded.to_csv("Dataset_encoded.csv", index=False)
print("âœ… One-hot encoding completed. Encoded file saved as Dataset_encoded.csv.")
```

```python
y=data[['Price']]
```

```python
x = pd.concat([data[['Total_area','Bedrooms','Bathrooms','Floors','Balcony']],
               encoded_df[['Location_Chennai','Location_Delhi','Location_Hyderabad','Location_Mumbai',
                           'Location_Pune','Parking_Yes','Property_type_Independent House','Property_type_Villa',
                           'Furnishing_status_Semi-Furnished','Furnishing_status_Unfurnished']]], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
Model=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
Model.fit(x_train, y_train.values.ravel())
```

```python
y_pred=Model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("The r2 score is:",r2)

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
print("The mean square error is:",mse)
```
